```python
cols_with_na = []
for col in df.columns:
    if df[col].isna().sum()>0:
        cols_with_na.append(col)
        
cols_with_na
```

```python
#outliers detection

plt.rcParams["figure.figsize"] = (9, 6)
df_list = ['AnnouncementsView','VisITedResources','raisehands']

fig, axes = plt.subplots(2,3)
fig.set_dpi(120)

count = 0
for r in range(2):
    for c in range(3):
        df[df_list[count].plot(kind = 'box',ax = axes[r,c])]
        count += 1
```
