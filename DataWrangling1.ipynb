```python
 import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("greenhouse-gas-emissions-industry-and-household-September-2022-quarter.csv")
```

```python
df = pd.DataFrame(data)
df
```

```python
#finding and replacing null values 
df.isnull()
```

```python
df['Data_value'] = df['Data_value'].fillna(df['Data_value'].mean())
df
```

```python
df['Variable'] = df['Variable'].map({'Seasonally adjusted' : 0, 'Actual' : 1}).astype(float)
df
```

```python
df = df[df['Data_value']>=12667]
df
```

```python
df = df.drop(['Units'],axis = 1)
```

```python
df
```

```python
#remove duplicate

non_duplicate = df[~df.duplicated('Anzsic_description')]
non_duplicate
```

```python
#factorize()

df['Period'] = pd.factorize(df['Period'])[0]
df
```

```python
#normalization

df.head()
```

```python
df.describe()
```

```python
sns.displot(df['Data_value'])
plt.show()
```

```python
df_copy = df.copy()
df_copy['Data_value'] = df_copy['Data_value']/df_copy['Data_value'].abs().max()
```

```python
df_copy = df.copy()
df_copy['Period'] = (df_copy['Period'] - df_copy['Period'].min())/(df_copy['Period'].max())
```

```python
sns.displot(df['Period'])
plt.show()
```

```python
z_copy = df.copy()
for col in ['Period']:
    z_copy[col] = (z_copy[col]-z_copy[col].mean())/z_copy[col].std()
```

```python
sns.displot(z_copy['Period'])
plt.show()
```
